```python
import pandas as pd
import json
```

```python
rb_pages = pd.read_csv('Data/Real Book Page Numbers - Sheet1.csv')
rb_pages = rb_pages.dropna()
rb1_pages = rb_pages[rb_pages['Book'] == 1]
rb2_pages = rb_pages[rb_pages['Book'] == 2]
rb1_pages.head()
```

```python
abc_dict = {}
for idx, row in rb1_pages.iterrows():
    if row['Song Name'][0].isalpha():
        folder = row['Song Name'][0]
    else:
        folder = "0-9"
    with open(f'Data/ABC_data/{folder}/{row['Song Name']}.abc', "r") as file:
        contents = file.read()
        contents.find('\n')

        for _ in range(2):
            idx = contents.find('V:')
            contents = contents[idx+4:]
        print(row['Song Name'])
        print(contents)
        abc_dict[row['Song Name']] = contents

for idx, row in rb2_pages.iterrows():
    if row['Song Name'][0].isalpha():
        folder = row['Song Name'][0]
    else:
        folder = "0-9"
    with open(f'Data/ABC_data/{folder}/{row['Song Name']}.abc', "r") as file:
        contents = file.read()
        contents.find('\n')

        for _ in range(2):
            idx = contents.find('V:')
            contents = contents[idx+4:]
        print(row['Song Name'])
        print(contents)
        abc_dict[row['Song Name']] = contents
```

```python
m_dict = {}

for title in abc_dict:
    measure_list = []

    song = abc_dict[title]
    while True:
        idx = song[2:].find('|')
        measure = song[:idx+1]
        song = song[idx+1:]

        if '\n' in measure:
            idx = measure.find('\n')
            measure = measure[idx+1:]
            
        if measure != ' ' and '%' not in measure:
            measure = measure.replace('|', '')
            measure = measure.replace('$', '')
            measure = measure.replace(':', '')
            while True:
                if measure[0] == ' ':
                    measure = measure[1:]
                else:
                    break
            measure_list.append(measure)
        if len(song) <= 9:
            break
    m_dict[title] = measure_list
```

```python
for title in m_dict:
    print(title)
    for i in m_dict[title]:
        print(i)
    print()
```

```python
with open('Data/measures.json', 'w') as f:
    json.dump(m_dict, f, indent=4)
```
