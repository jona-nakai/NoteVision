```python
import os
import subprocess
```

```python
xml_root = "XML_data"
abc_root = "ABC_data"
converter_script = "xml2abc.py"

for root, dirs, files in os.walk(xml_root):
    for filename in files:
        if filename.endswith(".xml"):
            in_path = os.path.join(root, filename)
            rel_path = os.path.relpath(in_path, xml_root)
            out_path = os.path.join(abc_root, os.path.splitext(rel_path)[0] + ".abc")
            
            os.makedirs(os.path.dirname(out_path), exist_ok=True)
            
            try:
                with open(out_path, 'w') as outfile:
                    subprocess.run(
                        ["python", converter_script, in_path],
                        stdout=outfile,
                        stderr=subprocess.PIPE,
                        check=True
                    )
                print(f"Converted {in_path} to {out_path}")
            except subprocess.CalledProcessError as e:
                print(f"Error processing {in_path}: {e.stderr.decode().strip()}")
```

```python
directory_path = 'ABC_Data'

file_count = 0
# Walk through all directories and files in the directory
for root, dirs, files in os.walk(directory_path):
    file_count += len(files)

file_count
```
