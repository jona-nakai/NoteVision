```python
import json
from pathlib import Path
import pandas as pd
import os
```

```python
with open('Data/measures.json', 'r') as f:
    measure_dict = json.load(f)

print(len(measure_dict))
```

```python
song = measure_dict['Afro-Blue']

song_abc = str()
for line in song:
    song_abc += line
    song_abc += ' | '
```

```python
data_map = pd.DataFrame(columns=['title', 'image_path', 'abc_measure'])

png_count = {}
abc_count = {}
correct = 0
incorrect = 0
for title in measure_dict:
    directory = Path(f'Data/measures_pngs/{title}/measures')
    file_count = len([f for f in directory.iterdir() if f.is_file()])
    png_count[title] = file_count
    abc_count[title] = len(measure_dict[title])
    print(title)
    print(f'png: {file_count}')
    print(f'abc: {len(measure_dict[title])}')
    if file_count == len(measure_dict[title]):
        print('correct')
        files = sorted(os.listdir(f'Data/measures_pngs/{title}/measures'))
        for i in range(len(measure_dict[title])):
            data_map.loc[len(data_map)] = [title, files[i], measure_dict[title][i]]
        correct += 1
    else:
        print('INCORRECT')
        incorrect += 1
    print()

print(f'correct: {correct}')
print(f'inccorect: {incorrect}')
```
